```python
new_cols_name= {
    'timestamp':'time',
    'cnt':'count',
    't1':'Temp_real_C',
    't2': 'Temp_feels_like_C',
    'hum': 'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}
```

```python
df.rename(new_cols_name,axis=1, inplace=True)
```

```python
df.head()
```

```python
## changing the humidity percent to value
df['humidity_percent']=df['humidity_percent']/100
```

```python
df.head()
```

```python
### converting the season integer values to the actual season representation
season_dict={
    '0.0':'spring',
    '1.0': 'summer',
    '2.0': 'autumn',
    '3.0': 'winter'
}


## converting the weather value to the actual weather representation

weather_dict= {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0': 'Broken clouds',
    '4.0': 'Cloudy',
    '7.0': 'Rain',
    '10.0': 'Rain with thunderstorm',
    '20.0': 'Snowfall'
}

## converting the season column data type to string
df.season=df.season.astype('str')

## mapping the season column data to the actual season
df.season=df.season.map(season_dict)

### converting the weather column data type to string
df.weather= df.weather.astype('str')
df.weather= df.weather.map(weather_dict)
```
